```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./Uncleaned_DS_jobs/Uncleaned_DS_jobs.csv')

# Drop rows with missing values
data = data.dropna()

# Extract numerical salary from 'Salary Estimate'
data = data[data['Salary Estimate'] != '-1']  # Remove rows with '-1' as salary estimate
data['Min Salary'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[0].replace('K','').replace('$','').replace(' ','')))
data['Max Salary'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[1].split('(')[0].replace('K','').replace('$','').replace(' ','')))
data['Avg Salary'] = (data['Min Salary'] + data['Max Salary']) / 2

# Group by company size and calculate salary statistics
salary_stats_by_size = data.groupby('Size')['Avg Salary'].describe()
print(salary_stats_by_size)

# Plot the salary distribution across different company sizes using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='Avg Salary', y='Size', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Company Size')
plt.title('Salary Distribution Across Different Company Sizes')
plt.show()

# Plot the salary distribution across different company sizes using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='Avg Salary', y='Size', data=data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Company Size')
plt.title('Salary Distribution Across Different Company Sizes')
plt.show()
```
