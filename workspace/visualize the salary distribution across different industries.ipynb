```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./Uncleaned_DS_jobs/Uncleaned_DS_jobs.csv')

# Drop rows with missing values
data = data.dropna()

# Extracting numerical salary from 'Salary Estimate'
data = data[data['Salary Estimate'] != '-1']  # Remove rows with '-1' as salary estimate
data['Min Salary'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[0].replace('K','').replace('$','').replace(' ','')))
data['Max Salary'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[1].split('(')[0].replace('K','').replace('$','').replace(' ','')))
data['Avg Salary'] = (data['Min Salary'] + data['Max Salary']) / 2

# Group by industry and calculate salary statistics
salary_stats_by_industry = data.groupby('Industry')['Avg Salary'].describe()
print(salary_stats_by_industry)

# Plot the salary distribution across industries using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='Avg Salary', y='Industry', data=data, hue='Industry', palette='viridis', dodge=False)
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Distribution Across Industries')
plt.legend().set_visible(False)  # Hide the legend
plt.show()

# Plot the salary distribution across industries using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='Avg Salary', y='Industry', data=data, hue='Industry', palette='viridis', dodge=False)
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Distribution Across Industries')
plt.legend().set_visible(False)  # Hide the legend
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('./Uncleaned_DS_jobs/Uncleaned_DS_jobs.csv')

# Drop rows with missing values
data = data.dropna()

# Extracting numerical salary from 'Salary Estimate'
data = data[data['Salary Estimate'] != '-1']  # Remove rows with '-1' as salary estimate
data['Min Salary'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[0].replace('K','').replace('$','').replace(' ','')))
data['Max Salary'] = data['Salary Estimate'].apply(lambda x: int(x.split('-')[1].split('(')[0].replace('K','').replace('$','').replace(' ','')))
data['Avg Salary'] = (data['Min Salary'] + data['Max Salary']) / 2

# Find the most frequent industries
top_industries = data['Industry'].value_counts().head(10).index
print(top_industries)

# Filter the dataset to include only the top 10 industries
top_industries_data = data[data['Industry'].isin(top_industries)]

# Group by industry and calculate salary statistics
salary_stats_by_industry = top_industries_data.groupby('Industry')['Avg Salary'].describe()
print(salary_stats_by_industry)

# Plot the salary distribution across top 10 industries using boxplot
plt.figure(figsize=(16, 10))
sns.boxplot(x='Avg Salary', y='Industry', data=top_industries_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Distribution Across Top 10 Industries')
plt.show()

# Plot the salary distribution across top 10 industries using violin plot
plt.figure(figsize=(16, 10))
sns.violinplot(x='Avg Salary', y='Industry', data=top_industries_data, palette='viridis')
plt.xlabel('Average Salary')
plt.ylabel('Industry')
plt.title('Salary Distribution Across Top 10 Industries')
plt.show()
```
